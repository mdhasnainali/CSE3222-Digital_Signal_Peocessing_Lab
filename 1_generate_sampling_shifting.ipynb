```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### Given Function:

y = 5 sin(2 * pi * 500t + 90)


```python
# Variables
start = 0
stop = 0.01
phase = 90
frequency = 500
amplitude = 5
sampling_frequency = 4000
samples_count = sampling_frequency * (stop-start)

# Conversion
phase_radian = (phase * np.pi/180)

# Base 
t_base = np.linspace(start=start, stop=stop, num=int(samples_count))
y_base = amplitude * np.sin((2 * np.pi * frequency * t_base) + phase_radian)
```

```python
# Generate Analog Signal
plt.figure(figsize=(10, 6))
plt.title(f'Continuous Signal')
plt.plot(t_base, y_base)
plt.show()
```

```python
# Generate Discrete Signal
plt.figure(figsize=(10, 6))
plt.title(f'Discrete Signal')
plt.stem(t_base, y_base)
plt.show()
```

```python
# Shifting Signal
shift = 2
t_shift = np.linspace(start=start, stop=stop, num=int(samples_count)) + shift
y_shift= amplitude * np.sin((2 * np.pi * frequency * t_shift) + phase_radian)
plt.figure(figsize=(10, 6))
plt.title(f'Discrete Signal Shifted by {shift}')
plt.stem(t_shift, y_shift)
plt.show()
```

```python
# Time Reversal of Signal
t_rev = - np.linspace(start=start, stop=stop, num=int(samples_count))
y_rev = amplitude * np.sin((2 * np.pi * frequency * t_rev) + phase_radian)
plt.figure(figsize=(10, 6))
plt.title(f'Reverse Discrete Signal')
plt.stem(t_rev, y_rev)
plt.stem(t_base, y_base)
plt.show()
```
