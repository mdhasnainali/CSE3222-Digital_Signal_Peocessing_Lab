```python
import numpy as np
import matplotlib.pyplot as plt
```

### Given Function:

y = 5 sin(2 * pi * 500t + 90)


```python
# Variables
start = 0
stop = 0.01
phase = 90
frequency = 500
amplitude = 5
sampling_frequency = 4000

# Conversion
phase_radian = (phase * np.pi/180)

# Base 
t_base = np.arange(start, stop, 1/sampling_frequency)
y_base = amplitude * np.sin((2 * np.pi * frequency * t_base) + phase_radian)
```

```python
# Generate Analog Signal
plt.figure(figsize=(10, 6))
plt.title(f'Continuous Signal')
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
plt.grid(True, which='both')
plt.axhline(color='r')
plt.axvline(color='r')
plt.plot(t_base, y_base)
plt.show()
```

```python
# Generate Discrete Signal
plt.figure(figsize=(10, 6))
plt.title(f'Discrete Signal')
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.grid(True, which='both')
plt.axhline(color='r')
plt.axvline(color='r')
plt.stem(y_base)
plt.show()
```

```python
# Shifting Signal
shift = 1
n = np.arange(-0.0010, 0.0010, 1/sampling_frequency)
y = amplitude * np.sin((2 * np.pi * frequency * n) + phase_radian)
n_shifted = n + shift

plt.figure(figsize=(10, 6))
plt.title(f'Discrete Signal Shifted by {shift} seconds')
plt.grid(True, which='both')
plt.axhline(color='r')
plt.axvline(color='r')

plt.stem(n, y, linefmt='b-', markerfmt='bo', basefmt=" ", label="Original Signal")
plt.stem(n_shifted, y, linefmt='g-', markerfmt='go', basefmt=" ", label="Shifted Signal")

plt.xlabel("n")
plt.ylabel("Amplitude")
plt.legend()
plt.show()
```

```python
# Time Reversal of Signal
n_rev = - np.arange(start, stop, 1/sampling_frequency)
y_rev = amplitude * np.sin((2 * np.pi * frequency * n_rev) + phase_radian)

plt.figure(figsize=(10, 6))
plt.title('Original and Reversed Discrete Signal')
plt.grid(True, which='both')
plt.axhline(color='r')
plt.axvline(color='r')

plt.stem(t_base, y_base, linefmt='b-', markerfmt='bo', basefmt=" ", label="Original Signal")
plt.stem(n_rev, y_rev, linefmt='g-', markerfmt='go', basefmt=" ", label="Reversed Signal")

plt.xlabel("n")
plt.ylabel("Amplitude")
plt.legend()
plt.show()
```
