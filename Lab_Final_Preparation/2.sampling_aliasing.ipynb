```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Basic

t = np.linspace(0, 0.008, 1000)
x = 5 * np.sin(2 * np.pi * 500 * t + np.radians(90))

plt.figure(figsize=(12, 12))
plt.subplots_adjust(hspace=0.3, wspace=0.2)

plt.subplot(3, 2, 1)
plt.plot(t, x)
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.title('x(n) = 5sin(2.500t + /2)')


# Sampling
fs = 4000
n = np.arange(0, 0.008, 1/fs)
x_n = 5 * np.sin(2 * np.pi * 500 * n + np.pi / 2)

plt.subplot(3, 2, 2)
plt.stem(x_n)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('x(n) = 5cos(2.(500/3750)n)')

# Up Sampling 
# Sampling Factor: 2
L = 2
upsampled_x = np.zeros(L * len(x_n))
upsampled_x[::2] = x_n

for i in range(1, len(upsampled_x)-1, 2):
    upsampled_x[i] = (upsampled_x[i-1] + upsampled_x[i+1]) / 2

plt.subplot(3, 2, 3)
plt.stem(upsampled_x)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('Upsampled')


# Down Sampling 
# Sampling Factor: 2
L = 2
downsampled_x = x_n[::L]

plt.subplot(3, 2, 4)
plt.stem(downsampled_x)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('Downsampled')
```
