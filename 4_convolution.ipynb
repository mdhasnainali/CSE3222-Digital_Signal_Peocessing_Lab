```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Convolution
def convolution(x, h):
    len_x = len(x)
    len_h = len(h)
    len_y = len_x + len_h - 1
    y = [0] * len_y

    for i in range(len_y):
        for k in range(len_x):
            if i-k >= 0 and i-k < len_h:
                y[i] += x[k] * h[i-k]
    return y
```

```python
x = [1,2,3,4,5]
h = [2,3,4]

y_conv = convolution(x,h)
y_corr = convolution(x,h[::-1])

print("Convolution: ", y_conv)
print("Correlation: ", y_corr)
```

```python
plt.figure(figsize=(16, 8))

plt.subplot(2, 3, 1)
plt.title(f'x(n)')
plt.stem(range(len(x)), x)

plt.subplot(2, 3, 2)
plt.title(f'h(n)')
plt.stem(range(len(h)), h)

plt.subplot(2, 3, 3)
plt.title(f'y(n) = x(n) * h(n)')
plt.stem(range(len(y_conv)), y_conv)

plt.subplot(2, 3, 4)
plt.title(f'x(n)')
plt.stem(range(len(x)), x)

plt.subplot(2, 3, 5)
plt.title(f'h(n)')
plt.stem(range(len(h)), h)

plt.subplot(2, 3, 6)
plt.title(f'y(n) = x(n) . h(n)')
plt.stem(range(len(y_corr)), y_corr)
```
