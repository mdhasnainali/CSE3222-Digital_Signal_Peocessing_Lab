```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Variables
start = 0
stop = 0.01
phase = 90
frequency = 500
amplitude = 5

# Conversion
phase_radian = (phase * np.pi/180)
```

### Given Function:

y = 5 sin(2 _ pi _ 500t + 90)


```python
# Base Signal
sampling_frequency = 4000
samples_count = sampling_frequency * (stop-start)

# Conversion
phase_radian = (phase * np.pi/180)

# Generate Discrete Signal
t_base = np.linspace(start=start, stop=stop, num=int(samples_count))
y_base = amplitude * np.sin((2 * np.pi * frequency * t_base) + phase_radian)
plt.figure(figsize=(10, 6))
plt.title(f'Base Signal')
plt.stem(t_base, y_base)
plt.show()
```

```python
# Up Sampling
sampling_frequency = 8000
samples_count = sampling_frequency * (stop-start)

# Conversion
phase_radian = (phase * np.pi/180)

# Generate Discrete Signal
t_up = np.linspace(start=start, stop=stop, num=int(samples_count))
y_up = amplitude * np.sin((2 * np.pi * frequency * t_up) + phase_radian)
plt.figure(figsize=(10, 6))
plt.title(f'Up Sampling')
plt.stem(t_up, y_up)
plt.show()
```

```python
# Down Sampling
sampling_frequency = 1000
samples_count = sampling_frequency * (stop-start)

# Conversion
phase_radian = (phase * np.pi/180)

# Generate Discrete Signal
t_down = np.linspace(start=start, stop=stop, num=int(samples_count))
y_down = amplitude * np.sin((2 * np.pi * frequency * t_down) + phase_radian)
plt.figure(figsize=(10, 6))
plt.title(f'Down Sampling')
plt.stem(t_down, y_down)
plt.show()
```

```python
# Generate Graph
plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
plt.title(f'Up Sampling (Sampling: 800Hz)')
plt.stem(t_up, y_up)

plt.subplot(1, 3, 2)
plt.title(f'Original Signal (Sampling: 400Hz)')
plt.stem(t_base, y_base)

plt.subplot(1, 3, 3)
plt.title(f'Down Sampling (Sampling: 100Hz)')
plt.stem(t_down, y_down)
```
