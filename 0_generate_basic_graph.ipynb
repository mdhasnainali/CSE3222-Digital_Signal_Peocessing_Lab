```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sine Wave
def sine_wave(amplitude, phase):
    phase_radian = (phase * np.pi/180)
    t = np.linspace(start = -100, stop = 100, num = 100)
    y_t = amplitude * np.sin(2 * np.pi * t + phase_radian) 
    plt.title(f'Sine Wave: A({amplitude}), P({phase} Degree)')
    plt.xlabel("Time(s)")
    plt.ylabel("Amplitude")
    plt.grid(True, which='both')
    plt.axhline(color='r')
    plt.axvline(color='r')
    plt.plot(t,y_t)
    plt.show()
```

```python
sine_wave(amplitude=2, phase=90)
```

```python
# Cosine Wave
def cosine_wave(amplitude, phase):
    phase_radian = (phase * np.pi/180)
    t = np.linspace(start = -100, stop = 100, num = 100)
    y_t = amplitude * np.cos(2 * np.pi * t + phase_radian) 
    plt.title(f'Cosine Wave: A({amplitude}), P({phase} Degree)')
    plt.xlabel("Time(s)")
    plt.ylabel("Amplitude")
    plt.grid(True, which='both')
    plt.axhline(color='r')
    plt.axvline(color='r')
    plt.plot(t,y_t)
    plt.show()
```

```python
cosine_wave(2, 0)
```

```python
def unit_step(start, end):
    n = np.arange(start,end)
    u = []
    
    for i in n:
        if i<0:
            u.append(0)
        else:
            u.append(1)
    
    plt.title(f'Unit Step: {start} to {end})')
    plt.xlabel("n")
    plt.ylabel("Amplitude")
    plt.grid(True, which='both')
    plt.axhline(color='r')
    plt.axvline(color='r')
    plt.stem(n,u)
    plt.grid()
    plt.show()
```

```python
unit_step(-5, 4)
```

```python
def ramp(start, end):
    n = np.arange(start,end)
    u_r = []
    
    for i in n:
        if i<0:
            u_r.append(0)
        else:
            u_r.append(i)
    
    plt.title(f'Ramp: {start} to {end})')
    plt.xlabel("n")
    plt.ylabel("Amplitude")
    plt.grid(True, which='both')
    plt.axhline(color='r')
    plt.axvline(color='r')
    plt.stem(n,u_r)
    plt.grid()
    plt.show()
```

```python
ramp(-5, 4)
```

```python
def exponential(start, end):
    n = np.arange(start,end)
    y_n = np.exp(n)
    
    plt.stem(n,y_n)
    plt.title(f'Exponential: {start} to {end})')
    plt.xlabel("n")
    plt.ylabel("Amplitude")
    plt.grid(True, which='both')
    plt.axhline(color='r')
    plt.axvline(color='r')
    plt.grid()
    plt.show()
```

```python
exponential(-3, 12)
```
