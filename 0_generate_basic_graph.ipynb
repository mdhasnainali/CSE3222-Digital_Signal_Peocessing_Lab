```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Sine Wave
def sine_wave(amplitude, phase):
    phase_radian = (phase * np.pi/180)
    x = np.linspace(start = -100, stop = 100, num = 100)
    y = amplitude * np.sin(2 * np.pi * x + phase_radian) 
    plt.title(f'Sine Wave: A({amplitude}), P({phase} Degree)')
    plt.grid(True, which='both')
    plt.axhline(color='r')
    plt.axvline(color='r')
    plt.plot(x,y)
    plt.show()
```

```python
sine_wave(amplitude=2, phase=90)
```

```python
# Cosine Wave
def cosine_wave(amplitude, phase):
    phase_radian = (phase * np.pi/180)
    x = np.linspace(start = -100, stop = 100, num = 100)
    y = amplitude * np.cos(2 * np.pi * x + phase_radian) 
    plt.title(f'Cosine Wave: A({amplitude}), P({phase} Degree)')
    plt.grid(True, which='both')
    plt.axhline(color='r')
    plt.axvline(color='r')
    plt.plot(x,y)
    plt.show()
```

```python
cosine_wave(2, 0)
```

```python
def unit_step(start, end, sampling_rate):
    x = np.linspace(start=start,stop=end,num=int((end-start)*sampling_rate) + 1)
    y = []
    
    for i in x:
        if i<0:
            y.append(0)
        else:
            y.append(1)
    
    plt.title(f'Unit Step: {start} to {end})')
    plt.grid(True, which='both')
    plt.axhline(color='r')
    plt.axvline(color='r')
    plt.stem(x,y)
    plt.grid()
    plt.show()
```

```python
unit_step(-5, 4, 3)
```

```python
def ramp(start, end, sampling_rate):
    x = np.linspace(start=start,stop=end,num=int((end-start)*sampling_rate) +1 )
    y = []
    
    for i in x:
        if i<0:
            y.append(0)
        else:
            y.append(i)
    
    plt.title(f'Ramp: {start} to {end})')
    plt.grid(True, which='both')
    plt.axhline(color='r')
    plt.axvline(color='r')
    plt.stem(x,y)
    plt.grid()
    plt.show()
```

```python
ramp(-5, 4, 3)
```

```python
def exponential(start, end, sampling_rate):
    x = np.linspace(start=start,stop=end,num=int((end-start)*sampling_rate) + 1)
    y = np.exp(x)
    
    plt.stem(x,y)
    plt.title(f'Exponential: {start} to {end})')
    plt.grid(True, which='both')
    plt.axhline(color='r')
    plt.axvline(color='r')
    plt.grid()
    plt.show()
```

```python
exponential(-3, 2, 3)
```
